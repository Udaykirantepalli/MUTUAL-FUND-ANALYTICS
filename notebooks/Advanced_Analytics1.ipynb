```python
nav = pd.read_csv(
    r"C:\Users\UDAY KIRAN\Desktop\mutual fund analytics\Data\processed\02_nav_history_clean.csv"
)

nav.head()
```

```python
nav = nav.sort_values(
    ["amfi_code","date"]
)

nav["daily_return"] = (
    nav.groupby("amfi_code")["nav"]
       .pct_change()
)

nav.head()
```

```python
var_cvar_results = []

for code in nav["amfi_code"].unique():

    fund = nav[
        nav["amfi_code"] == code
    ]

    returns = fund["daily_return"].dropna()

    if len(returns) > 30:

        var_95 = returns.quantile(0.05)

        cvar_95 = returns[
            returns <= var_95
        ].mean()

        var_cvar_results.append(
            [
                code,
                var_95,
                cvar_95
            ]
        )

var_cvar_df = pd.DataFrame(
    var_cvar_results,
    columns=[
        "amfi_code",
        "VaR_95",
        "CVaR_95"
    ]
)

var_cvar_df.head()
```

```python
transactions = pd.read_csv(
    r"C:\Users\UDAY KIRAN\Desktop\mutual fund analytics\Data\processed\08_investor_transactions_clean.csv"
)

transactions.head()
```

```python
transactions["transaction_date"] = pd.to_datetime(
    transactions["transaction_date"]
)
```

```python
first_txn = (
    transactions.groupby("investor_id")
    ["transaction_date"]
    .min()
)

transactions["cohort_year"] = (
    transactions["investor_id"]
    .map(first_txn.dt.year)
)

transactions.head()
```

```python
transactions.head()
```

```python
cohort_summary = (
    transactions.groupby("cohort_year")
    .agg(
        avg_sip_amount=("amount_inr","mean"),
        total_invested=("amount_inr","sum")
    )
    .reset_index()
)

cohort_summary
```

```python
top_fund = (
    transactions.groupby(
        ["cohort_year","amfi_code"]
    )
    .size()
    .reset_index(name="count")
)

top_fund = top_fund.sort_values(
    ["cohort_year","count"],
    ascending=[True,False]
)

top_fund.groupby("cohort_year").head(1)
```

```python
sip_only = transactions[
    transactions["transaction_type"] == "SIP"
].copy()

sip_only = (
    sip_only.groupby("investor_id")
    .filter(lambda x: len(x) >= 6)
)
```

```python
sip_only = sip_only.sort_values(
    ["investor_id","transaction_date"]
)

sip_only["gap_days"] = (
    sip_only.groupby("investor_id")
    ["transaction_date"]
    .diff()
    .dt.days
)
```

```python
gap_summary = (
    sip_only.groupby("investor_id")
    ["gap_days"]
    .mean()
    .reset_index()
)

gap_summary.head()
```

```python
gap_summary["status"] = np.where(
    gap_summary["gap_days"] > 35,
    "At Risk",
    "Healthy"
)

gap_summary.head()
```

```python
gap_summary["status"].value_counts()
```

```python
gap_summary["status"].value_counts()
```

```python
total = len(gap_summary)

healthy_pct = (
    (gap_summary["status"]=="Healthy")
    .sum()
    / total
) * 100

risk_pct = (
    (gap_summary["status"]=="At Risk")
    .sum()
    / total
) * 100

print("Healthy:", round(healthy_pct,2), "%")
print("At Risk:", round(risk_pct,2), "%")
```

```python
portfolio = pd.read_csv(
    r"C:\Users\UDAY KIRAN\Desktop\mutual fund analytics\Data\processed\09_portfolio_holdings_clean.csv"
)

portfolio.head()
```

```python
portfolio.columns
```

```python
portfolio["weight_sq"] = (
    portfolio["weight_pct"] / 100
) ** 2
```

```python
hhi_df = (
    portfolio.groupby("amfi_code")
    ["weight_sq"]
    .sum()
    .reset_index()
)

hhi_df.head()
```

```python
hhi_df["concentration"] = np.where(
    hhi_df["weight_sq"] > 0.25,
    "Highly Concentrated",
    np.where(
        hhi_df["weight_sq"] > 0.15,
        "Moderately Concentrated",
        "Diversified"
    )
)

hhi_df.head()
```
