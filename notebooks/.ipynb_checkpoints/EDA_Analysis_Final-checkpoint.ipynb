```python
category = pd.read_csv(
    r"C:\Users\UDAY KIRAN\Desktop\mutual fund analytics\Data\PROCESSED\05_category_inflows_clean.csv"
)

print(category.head())
```

```python
heatmap_data = category.pivot(
    index="category",
    columns="month",
    values="net_inflow_crore"
)

heatmap_data.head()
```

```python
plt.figure(figsize=(16,8))

sns.heatmap(
    heatmap_data,
    cmap="YlGnBu",
    annot=False
)

plt.title("Category-wise Net Inflows Heatmap")
plt.tight_layout()

plt.show()
```

```python
plt.figure(figsize=(16,8))

sns.heatmap(
    heatmap_data,
    cmap="YlGnBu"
)

plt.title("Category-wise Net Inflows Heatmap")

plt.savefig(
    r"C:\Users\UDAY KIRAN\Desktop\mutual fund analytics\charts\category_heatmap.png",
    bbox_inches="tight"
)

plt.show()

print("Heatmap Saved Successfully!")
```

```python
investor = pd.read_csv(
    r"C:\Users\UDAY KIRAN\Desktop\mutual fund analytics\Data\PROCESSED\08_investor_transactions_clean.csv"
)

investor.head()
```

```python
age_counts = investor["age_group"].value_counts()

plt.figure(figsize=(8,8))

plt.pie(
    age_counts,
    labels=age_counts.index,
    autopct="%1.1f%%"
)

plt.title("Investor Age Group Distribution")

plt.savefig(
    r"C:\Users\UDAY KIRAN\Desktop\mutual fund analytics\charts\age_group_pie.png",
    bbox_inches="tight"
)

plt.show()

print("Age Group Pie Saved!")
```

```python
gender_counts = investor["gender"].value_counts()

plt.figure(figsize=(8,8))

plt.pie(
    gender_counts,
    labels=gender_counts.index,
    autopct="%1.1f%%"
)

plt.title("Investor Gender Distribution")

plt.savefig(
    r"C:\Users\UDAY KIRAN\Desktop\mutual fund analytics\charts\gender_split.png",
    bbox_inches="tight"
)

plt.show()

print("Gender Chart Saved!")
```

```python
plt.figure(figsize=(12,6))

sns.boxplot(
    data=investor,
    x="age_group",
    y="amount_inr"
)

plt.title("Investment Amount Distribution by Age Group")

plt.xticks(rotation=45)

plt.savefig(
    r"C:\Users\UDAY KIRAN\Desktop\mutual fund analytics\charts\amount_boxplot.png",
    bbox_inches="tight"
)

plt.show()

print("Box Plot Saved!")
```

```python
state_amount = investor.groupby("state")["amount_inr"].sum().sort_values()

plt.figure(figsize=(12,8))

state_amount.plot(kind="barh")

plt.title("Total Investment Amount by State")
plt.xlabel("Amount (INR)")

plt.savefig(
    r"C:\Users\UDAY KIRAN\Desktop\mutual fund analytics\charts\state_distribution.png",
    bbox_inches="tight"
)

plt.show()

print("State Distribution Saved!")
```

```python
tier_counts = investor["city_tier"].value_counts()

plt.figure(figsize=(8,8))

plt.pie(
    tier_counts,
    labels=tier_counts.index,
    autopct="%1.1f%%"
)

plt.title("T30 vs B30 Investor Distribution")

plt.savefig(
    r"C:\Users\UDAY KIRAN\Desktop\mutual fund analytics\charts\t30_b30_pie.png",
    bbox_inches="tight"
)

plt.show()

print("T30/B30 Chart Saved!")
```

```python
performance = pd.read_csv(
    r"C:\Users\UDAY KIRAN\Desktop\mutual fund analytics\Data\PROCESSED\07_scheme_performance_clean.csv"
)

performance.head()
```

```python
corr_columns = [
    "return_1yr_pct",
    "return_3yr_pct",
    "return_5yr_pct",
    "alpha",
    "beta",
    "sharpe_ratio"
]

corr = performance[corr_columns].corr()

plt.figure(figsize=(10,8))

sns.heatmap(
    corr,
    annot=True,
    cmap="coolwarm"
)

plt.title("Fund Performance Correlation Matrix")

plt.savefig(
    r"C:\Users\UDAY KIRAN\Desktop\mutual fund analytics\charts\correlation_matrix.png",
    bbox_inches="tight"
)

plt.show()

print("Correlation Matrix Saved!")
```

```python
holdings = pd.read_csv(
    r"C:\Users\UDAY KIRAN\Desktop\mutual fund analytics\Data\PROCESSED\09_portfolio_holdings_clean.csv"
)

holdings.head()
```

```python
sector = holdings.groupby("sector")["weight_pct"].sum()

plt.figure(figsize=(8,8))

plt.pie(
    sector,
    labels=sector.index,
    autopct="%1.1f%%"
)

centre_circle = plt.Circle((0,0),0.70,fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)

plt.title("Sector Allocation Across Equity Funds")

plt.savefig(
    r"C:\Users\UDAY KIRAN\Desktop\mutual fund analytics\charts\sector_donut.png",
    bbox_inches="tight"
)

plt.show()

print("Sector Donut Saved!")
```
